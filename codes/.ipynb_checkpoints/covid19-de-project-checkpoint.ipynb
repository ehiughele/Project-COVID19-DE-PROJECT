```python
import time
import pandas as pd

def download_and_load_query_results(
    client: boto3.client, query_response: Dict, timeout_seconds: int = 30
) -> pd.DataFrame:
    start_time = time.time()
    while True:
        try:
            # Increase the limit to retrieve more rows in each call (e.g., 10000)
            result = client.get_query_results(
                QueryExecutionId=query_response["QueryExecutionId"],
                MaxResults=1000,
            )
            # Check if the query execution is completed
            if result.get("QueryExecution", {}).get("Status", {}).get("State") == "SUCCEEDED":
                break
        except Exception as err:
            if "not yet finished" in str(err):
                time.sleep(0.001)
            elif time.time() - start_time >= timeout_seconds:
                raise TimeoutError("Query execution timed out.") from err
            else:
                raise err
    
    temp_file_location: str = "athena_query_results"
    output_location = query_response["ResultConfiguration"]["OutputLocation"]
    file_extension = output_location.split(".")[-1].lower()
    
    if file_extension == "csv":
        temp_file_location += ".csv"
    elif file_extension == "json":
        temp_file_location += ".json"
    else:
        raise ValueError(f"Unsupported file extension: {file_extension}")
    
    s3_client = boto3.client(
        "s3",
        aws_access_key_id=AWS_ACCESS_KEY,
        aws_secret_access_key=AWS_SECRET_KEY,
        region_name=AWS_REGION,
    )
    s3_client.download_file(
        S3_BUCKET_NAME,
        output_location,
        temp_file_location,
    )
    
    if file_extension == "csv":
        return pd.read_csv(temp_file_location)
    elif file_extension == "json":
        return pd.read_json(temp_file_location)
```

```python
response = athena_client.start_query_execution(
    QueryString="SELECT * FROM enigma_jhud",
    QueryExecutionContext={"Database":SCHEMA_NAME},
    ResultConfiguration={
            "OutputLocation":S3_STAGING_DIR,
            "EncryptionConfiguration": {"EncryptionOption": "SSE_S3"},
    },
)
enigma_jhud = download_and_load_query_results(athena_client, response)
```
